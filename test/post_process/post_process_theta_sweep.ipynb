```python
import pandas as pd
df = pd.read_csv("../theta_sweep.csv")
display(df)
```

```python
df_fall = df[df["PULSE_POS"]=="FALL"].reset_index(drop=True)
df_fall["HW_OUT_INV"] = pd.to_numeric(df_fall["HW_OUT_INV"])
df_fall.plot(x="THETA_PS", y="HW_OUT_INV", marker = 'o', title='Falling Transition w/ Tclk=40ns')

df_rise = df[df["PULSE_POS"]=="RISE"].reset_index(drop=True)
df_rise["HW_OUT"] = pd.to_numeric(df_rise["HW_OUT"])
df_rise.plot(x="THETA_PS", y="HW_OUT", marker = 'o', title='Rising Transition  w/ Tclk=40ns')
```
